```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('files/1/fifa.csv')
```

```python
data.head()
```

```python
pd.set_option('display.max_rows', 164)
data['Nationality'].value_counts()
```

```python
def decompose_categorical(x):
    if x == 'England':
        return 'England'
    if x == 'Germany':
        return 'Germany'
    if x == 'Spain':
        return 'Spain'
    if x == 'Argentina':
        return 'Argentina'
    if x == 'France':
        return 'France'
    
    return 'Others'
        
data['Nationality'] = data['Nationality'].apply(decompose_categorical)
data['Nationality'].value_counts()
```
