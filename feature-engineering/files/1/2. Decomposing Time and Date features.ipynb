### Decomposing Date and Time features

Date variables are a special type of categorical variable.

From a date variable, we can extract basic features like:

Year

Month

Day

Day of the week

Time difference

```python
#Importing important libraries

import pandas as pd
import numpy as np
import datetime
```

```python
#Creating a time series. Periods =15 means it creates 15 rows

time_series = pd.date_range('1/1/2020', periods=15)

df = pd.DataFrame({'date':time_series})
df
```

```python
df.dtypes
```

```python
#Converting to DateTime type

df['date']=pd.to_datetime(df['date'])
```

```python
#Extracting Year, Month and Day from the data

df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
df['day']=df['date'].dt.day
```
